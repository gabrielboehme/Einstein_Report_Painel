```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import urllib.request, urllib.parse
import html5lib
```

```python
df = pd.read_csv(r'ETL/hospital_capacity_by_us_state.csv')
covid_data = pd.read_json('https://covidtracking.com/api/v1/states/current.json').sort_values('state')
covid_data = covid_data[['state','death','hospitalized','positive']]
covid_data
```

```python
labels = df.columns[0].split(';')
df = df.iloc[:52,0].str.split(';',expand=True)
df.columns = labels
df_covid = df.merge(covid_data,how='left',left_on='state',right_on='state')
for column in df_covid.columns:
    df_covid[column] = df_covid[column].astype('int64',errors='ignore')
df_covid
```

```python
corr = df_covid.corr()
corr.style.background_gradient(cmap='coolwarm')
```
